# Model

```python
x_train=t_train.drop(['Survived'],axis=1)
y_train=t_train['Survived']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
params={'class_weight':['balanced',None],
        'penalty':['l1','l2'],
        'C':np.linspace(0.01,1000,10)}
```

```python
model=LogisticRegression(fit_intercept=True)
```

```python
grid_search=GridSearchCV(model,cv=20,param_grid=params,scoring='roc_auc')
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_estimator=grid_search.best_estimator_
```

```python
best_estimator.fit(x_train,y_train)
```

```python
predictions=best_estimator.predict(t_test)
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```

```python
report(grid_search.cv_results_,3)
```
